```python
from mpl_toolkits import mplot3d 
import numpy as np 
import matplotlib.pyplot as plt 

print("setup complte")

#creating dataset

with open('C:/Users/HP/Desktop/MTECH 1ST SEM NOTES/PROJECT BY OWN/src/object sensitivity vs risk simulation/normals_view.txt', "r") as f1:
    lines1 = f1.readlines()
    x1 = [float(line.split(",")[9]) for line in lines1]
    z1 = [float(line.split(",")[8]) for line in lines1]
    y1 = [float(line.split(",")[1]) for line in lines1]
    
#creating figure

fig = plt.figure(figsize = (10, 7)) 
ax = plt.axes(projection ="3d") 


# Creating plot 
plt.xlabel("object_sensitivity")
plt.ylabel("SSL")
plt.zlabel("risk_score")
ax.scatter3D(x1, y1, z1, color = "green", label = ("normal-view")); 
plt.title("risk score on SSL and object sensitivity")

plt.show()
```
