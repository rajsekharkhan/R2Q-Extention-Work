```python
import numpy as np
import matplotlib.pyplot as plt

def inverse_gompertz(mu_score,b,c):
    mu_1 = mu_score * 100
    a = 1
    
    
    r1 = c*mu_1
    r1 = (-1)*r1
    
    m1= b*np.exp(r1)
    m1=(-1)*m1
   
    reputation_score_calculated = (1-(a* np.exp(m1)));
    return reputation_score_calculated

t1 = np.arange(0.0,1.01,0.05)

plt.figure(figsize=(8,8))
plt.xlabel("unalikability score")
plt.ylabel("reputation score")
plt.title("inverse gompertz function(1)")

plt.plot(t1,inverse_gompertz(t1,50,0.1),'b-',label ="B=50, C=0.1")
plt.plot(t1,inverse_gompertz(t1,150,0.1), "r-", label ='B=150, C=0.1')
plt.plot(t1,inverse_gompertz(t1,300,0.1), "g-", label ='B=300, C=0.1')
plt.plot(t1,inverse_gompertz(t1,500,0.1), "y-", label ='B=500, C=0.1')
plt.legend()
plt.show()
    
```
