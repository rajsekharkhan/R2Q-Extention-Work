```python
import numpy as np
import matplotlib.pyplot as plt

def uncertainty_measure(reputation,lamda):
    for n in reputation:
        if n == 0 :
            co_eff_SSL = 1
        else :
            r1 = lamda/reputation
            r1 = -1 * r1
            m1 = np.exp(r1)
            co_eff_SSL = 1-m1
    print(co_eff_SSL)
    return co_eff_SSL
        
t1 = np.arange(0.0,1.01,0.2)

plt.figure(figsize=(12,8))
plt.xlabel("Reputation Score")
plt.ylabel("uncertainty(wl)")
plt.title("co_efficient of SSL")

plt.plot(t1,uncertainty_measure(t1,0.25),"b-", label="lamda = 0.25")
plt.plot(t1,uncertainty_measure(t1,0.45),"r-", label="lamda = 0.45")
plt.plot(t1,uncertainty_measure(t1,0.65),"g-", label="lamda = 0.65")

plt.legend()
plt.show()



```
