```python
import numpy as np
import matplotlib.pyplot as plt

def ws_measure(expected_damage,v) :
    m1 = (expected_damage/v)
    m1 = -1*m1
    r1 = np.exp(m1)
    co_eff_obj_sens = 1-r1
    
    return co_eff_obj_sens

t1 = np.arange(0,1.01,0.05)

plt.figure(figsize=(10,8))
plt.xlabel("expected damage")
plt.ylabel("utility(ws)")
plt.title("Ws measure")

plt.plot(t1,ws_measure(t1,0.6),"b-",label="v=0.6")
plt.plot(t1,ws_measure(t1,0.8),"r-",label="v=0.8")
plt.plot(t1,ws_measure(t1,1.0),"g-",label="v=1.0")
plt.legend()
plt.show()
```
