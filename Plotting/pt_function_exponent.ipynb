```python
import numpy as np
import matplotlib.pyplot as plt


def pt_function(threat,phi,gamma) : 
    risk = list()
    for i in range(len(threat)) :
        if threat[i] >= 0.5 :
             risk.append(np.power(threat[i],phi))
        else :
             risk.append(np.power(threat[i],gamma))
            
    print(risk)
    return risk

plt.figure(figsize=(10,8))
plt.xlabel("expected threat")
plt.ylabel("risk")
plt.title("PT function exponent")

t1 = np.arange(0.0,1.01,0.05)

plt.plot(t1, pt_function(t1,0.2,1.15), "r--", label="phi=0.2, gamma=1.15")
plt.plot(t1, pt_function(t1,0.3,1.25), "b--", label="phi=0.3, gamma=1.25")
plt.plot(t1, pt_function(t1,0.4,1.35), "g--", label="phi=0.4, gamma=1.35")

plt.legend()
plt.show()
```
